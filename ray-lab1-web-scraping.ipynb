# Lab 1 : Scraping Single Page

```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url1 = "https://www.popvortex.com/music/charts/top-100-songs.php"
response1 = requests.get(url1)
response1.status_code # 200 status code means OK!
```

```python
response1.content
```

```python
soup1 = BeautifulSoup(response1.content, "html.parser")
```

```python
soup1.find_all('p', class_='title-artist')
```

```python
soup_container1 = soup1.find_all('p', class_='title-artist')

soup_data1 = []
for container in soup_container1:
    song = container.find('cite', class_='title').text.strip()
    artist = container.find('em', class_='artist').text.strip()
    soup_data1.append({'Artist': artist, 'Song': song})

#print(f"Artist: {item['Artist']}, Song: {item['Song']}")
    
soup_data1_df = pd.DataFrame(soup_data1)
```

```python
display(soup_data1_df)
```

# Lab 2: Continuation of Lab 1

```python
# Alternative source for top songs

url2 = "https://kworb.net/spotify/songs_2023.html"
response2 = requests.get(url2)
response2.status_code
```

```python
soup2 = BeautifulSoup(response2.content, "html.parser")
```

```python
print(soup2.prettify())
```

```python
# select the relevant html container where the data is

soup2.select("body > div.container > div:nth-child(4) > table > tbody > tr")
```

```python
artist2 = []
title2 = []

soup_container2 = soup2.select("body > div.container > div:nth-child(4) > table > tbody > tr")

for container2 in soup_container2:
    div_content = container2.find('td', class_='text').text.strip()
    artist, title = div_content.split(" - ", 1)
    artist2.append(artist.strip())
    title2.append(title.strip())

print(artist2)
print(title2)
```

```python
soup_data2_df = pd.DataFrame({"Artist":artist2,
                           "Song":title2
                          })

display(soup_data2_df)
```

```python
#Concat the two DFs together

soup_music_df = pd.concat([soup_data1_df, soup_data2_df], ignore_index=True)
display(soup_music_df)
```
