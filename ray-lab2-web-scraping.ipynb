```python
# USA Codes
```

```python
url1 = "http://uscode.house.gov/download/download.shtml"
response1 = requests.get(url1)
soup1 = BeautifulSoup(response1.text, 'html.parser')
```

```python
#print(soup1.find('div', class_='uscitemlist'))
```

```python
# I use the .find() instead of .select() here because it seems the html code has a clear structure

uscitemlist_div = soup1.find('div', class_='uscitemlist')
changed_text = uscitemlist_div.find_all('div',class_='usctitlechanged')

changed_list = [changes.get_text(strip=True) for changes in changed_text]
print(changed_list)
```

```python
# UK Data Website
```

```python
url3 = "https://www.data.gov.uk/"
response3 = requests.get(url3)
soup3 = BeautifulSoup(response3.text, 'html.parser')
```

```python
print(soup3.find('div', class_='govuk-grid-column-full'))
```

```python
categories_div = soup3.find('div', class_='govuk-grid-column-full')
h3_elements = categories_div.find_all('h3')

h3_values = [element.get_text(strip=True) for element in h3_elements]
display(h3_values)
```
